```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import folium
import os
```

```python
from google.colab import files
uploaded = files.upload()
data=pd.read_csv(io.BytesIO(uploaded['CAUSES_OF_DEATH.csv']))
```

```python
#importing shapefiles for the map
uploaded = files.upload()

file_extensions = ['.shp', '.dbf', '.prj', '.shx']

base_filename = os.path.splitext(list(uploaded.keys())[0])[0]

for ext in file_extensions:
    file_path = base_filename + ext
    if file_path not in uploaded:
        raise ValueError(f"Required file {file_path} is missing.")

shapefile_path = base_filename + '.shp'
geo_data = gpd.read_file(shapefile_path, encoding='latin1')
```

```python
df = data[['Country_name', 'Outdoor_air_pollution', 'High_blood_pressure', 'Alcohol_use', 'Smoking','Drug_use']]

#these are not countries, they are affecting the map, so I removed them
countries_to_remove = ['World','G20','World Bank Upper Middle Income','World Bank Lower Middle Income','East Asia & Pacific (WB)','Western Pacific Region (WHO)']
df = df[~df['Country_name'].isin(countries_to_remove)]
```

```python
df['Outdoor_air_pollution'] = pd.to_numeric(df['Outdoor_air_pollution'])

m = folium.Map(location=[0, 0], zoom_start=2)

folium.Choropleth(
    geo_data=geo_data,
    name='choropleth',
    data=df,
    columns=['Country_name', 'Outdoor_air_pollution'],
    key_on='feature.properties.SOVEREIGNT',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Number of Deaths by Outdoor_air_pollution',
    reset=True,
    bins=[0, 500, 1000, 2500, 5000, 7500, 10000, 15000, 20000, 25000, 50000, 100000, 1000000, float('inf')] 
).add_to(m)

m.save('outdoor_deaths_map.html')

from google.colab import files
files.download('outdoor_deaths_map.html')
```
