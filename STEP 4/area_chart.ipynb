```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('min.csv')

df_sorted = df.sort_values('min(Matterpollution)', ascending=True)
top_5_countries = df_sorted.head(5)

plt.stackplot(top_5_countries['Entity'],
              top_5_countries['min(Matterpollution)'],
              top_5_countries['min(OzonePollution)'],
              top_5_countries['min(concentration)'],
              labels=['min(Matterpollution)', 'min(OzonePollution)', 'min(concentration)'])

plt.xlabel('Countries')
plt.ylabel('Pollution Level')

plt.title('Min Pollution Levels of Top 5 Countries')

plt.legend()

plt.show()
```
