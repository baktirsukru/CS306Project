```python
import mysql.connector
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="1234",
  database="outdoor_pollution"
)

mycursor = mydb.cursor()
```

```python
query="SELECT country_name, year, percentage FROM percentage_under_4 INNER JOIN countries ON countries.country_id = percentage_under_4.country_id"

mycursor.execute(query)
```

```python
df = pd.DataFrame(mycursor.fetchall())
```

```python
df.columns = ["country_name", "year", "percentage"]
```

```python
ang = df.loc[df["country_name"] == "Angola"]
arg = df.loc[df["country_name"] == "Argentina"]
berm = df.loc[df["country_name"] == "Bermuda"]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10,8))
ang.plot.scatter(x='year', y='percentage', c='DarkBlue',label='Angola', ax=ax1)
arg.plot.scatter(x='year', y='percentage', c='DarkBlue',label='Argentina', ax=ax2)
berm.plot.scatter(x='year', y='percentage', c='DarkBlue',label='Bermuda', ax=ax3)
plt.legend()
plt.show()
```
